```python
import pandas as pd
import glob


```

```python
# Cargar los archivos CSV en un DataFrame
df = pd.concat(map(pd.read_csv, glob.glob("../../data/raw/*.csv")))
df.columns
```

```python
# Cambiar el nombre de las columnas a minúsculas
df.columns = df.columns.str.lower()
df.columns
```

```python
# Reemplazar espacios y caracteres especiales por guiones bajos
df.columns = df.columns.str.replace(r'\W+', '_', regex=True)
df.columns
```

```python
#indice de las columnas
[i for i, col in enumerate(df.columns) if col == "fecha_arribo"]
```

```python
# Renombrar columnas duplicadas
cols = pd.Series(df.columns)
for dup in cols[cols.duplicated()].unique():
    cols[cols[cols == dup].index] = [
        f"{dup}_{i}" if i != 0 else dup
        for i in range(sum(cols == dup))
    ]

df.columns = cols
```

```python
# combinar las columnas "fecha_arribo" y "fecha_arribo_1" en una nueva columna "fecha_arribo_final"
df["fecha_arribo_final"] = df["fecha_arribo"].combine_first(df["fecha_arribo_1"])
```

```python
# eliminar las columnas originales "fecha_arribo" y "fecha_arribo_1"
df.drop(columns=["fecha_arribo", "fecha_arribo_1"], inplace=True)
```

```python
df.columns
```

```python
df["edad_usuario"] = df["edad_usuario"].astype("Int64")
```

```python
#Cambiar el tipo de dato de las columnas "bici", "ciclo_estacion_retiro" y "ciclo_estacionarribo" a enteros, manejando los errores con "coerce" para convertir los valores no convertibles en NaN, y luego convertir la serie resultante a tipo "Int64" para permitir valores enteros con NaN.
cols = ["bici","ciclo_estacion_retiro","ciclo_estacionarribo"]


df[cols] = df[cols].apply(pd.to_numeric, errors="coerce").astype("Int64")
```

```python
df["fecha_arribo_final"] = pd.to_datetime(
    df["fecha_arribo_final"],
    dayfirst=True,
    errors="coerce"
)
```
