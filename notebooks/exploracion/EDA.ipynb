## Limpieza de Datos

```python
df.info()

```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["genero_usuario"].unique()
```

```python
df["genero_usuario"] = df["genero_usuario"].str.strip().str.lower()
```

```python
df["genero_usuario"].unique()
```

```python
df["genero_usuario"] = (
    df["genero_usuario"]
    .replace({
        "m": "masculino",
        "f": "femenino",
        "o": "otro",
        "?": "desconocido"
    })
    .fillna("desconocido")
)
```

```python
df["genero_usuario"].value_counts()
```

```python
df["genero_usuario"].unique()
```

```python
df["edad_usuario"].unique()
```

```python
df["edad_usuario"].describe()
```

```python
df.isnull().sum()
```

```python
# eliminar edades menores a 16 y mayores a 75
df = df[(df["edad_usuario"] >= 16) & (df["edad_usuario"] <= 75)]
```

```python
Q1 = df["edad_usuario"].quantile(0.25)
Q3 = df["edad_usuario"].quantile(0.75)


IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

df = df[
    (df["edad_usuario"] >= limite_inferior) &
    (df["edad_usuario"] <= limite_superior)
]
```

```python
df["edad_usuario"].hist(bins=30)
```

```python
df["ciclo_estacion_retiro"].unique()
```

```python
# Eliminar filas con valores nulos en ciclo_estacion_retiro
df = df.dropna(subset=["ciclo_estacion_retiro"])
```

```python
# eliminar ciclo_estacionarribo
df = df.dropna(subset=["ciclo_estacionarribo"])
```

```python
df.isnull().sum()
```
