### Formato correcto 

```python
cols = ["bici","ciclo_estacion_retiro","ciclo_estacionarribo"]

df[cols] = df[cols].apply(pd.to_numeric, errors="coerce").astype("Int64")
```

```python
df['hora_arribo'] = pd.to_datetime(df['hora_arribo'], format='%H:%M:%S')
```

```python
df['hora_retiro'] = pd.to_datetime(df['hora_retiro'], format='%H:%M:%S')
```

```python
# porcentaje de valores nulos por columna
df.tail()
```

```python
#genero_usuario tiene 0, pero edad_usuario tiene 1, entonces se eliminará esa fila
df.isnull().sum()
```

```python
# porcentaje de valores nulos por columna
print(df.isnull().mean() * 100)
```

```python
# Se eliminarán las filas con valores nulos en las columnas críticas para el análisis
columnas_criticas = ['ciclo_estacionarribo','ciclo_estacion_retiro','edad_usuario','genero_usuario']
df.dropna(subset=columnas_criticas, inplace=True)
```

```python
df.isnull().sum()
```

```python
df['fecha'] =  df['hora_arribo'] - df['hora_retiro']

df['fecha']
```

```python
# Convierte cualquier diferencia negativa en positiva
df['fecha'] = df['fecha'].abs()

df['fecha']
```
