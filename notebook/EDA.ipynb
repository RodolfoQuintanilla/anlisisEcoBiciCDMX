```python
#separa dia y hora
df['hora'] = df['fecha'].str.split(' ').str[-1]
```

```python
df['fecha'] = pd.to_timedelta(df['fecha'])
```

```python
#df['tiempo_hora'] = df['fecha'].dt.strftime('%H:%M:%S')
df['horas_sin_dias'] = (df['fecha'] % pd.Timedelta(days=1)).dt.total_seconds() / 3600
```

```python
df['minutos'] = df['horas_sin_dias'] * 60
```

### Analizar por qué hay registros mayores a 50 minutos y

```python
#Detectar outliers
q1 = df['minutos'].quantile(0.25)
q3 = df['minutos'].quantile(0.75)
iqr = q3 - q1

limite_superior = q3 + 1.5 * iqr

outliers = df[df['minutos'] > limite_superior]
```

```python
# usuarios > 50 min
df_mayor_50 = df[df['minutos'] > 50]

# posibles errores extremos
df_error = df[df['minutos'] > 300]

# porcentaje
porcentaje = len(df_mayor_50) / len(df) * 100

print(f"Usuarios > 50 min: {len(df_mayor_50)} ({porcentaje:.2f}%)")
print(f"Posibles errores (>300 min): {len(df_error)}")
```

## Determinar la duración media por trayecto para establecer un "estándar" de uso.

```python
mediaDuracion = df['minutos'].mean()
print(f'La duracion media de los viajes es: {mediaDuracion:.2f} minutos')
```
