```python
#separa dia y hora
df['hora'] = df['fecha'].str.split(' ').str[-1]
```

```python
df['fecha'] = pd.to_timedelta(df['fecha'])
```

```python
#df['tiempo_hora'] = df['fecha'].dt.strftime('%H:%M:%S')
df['horas_sin_dias'] = (df['fecha'] % pd.Timedelta(days=1)).dt.total_seconds() / 3600
```

```python
df['minutos'] = df['horas_sin_dias'] * 60
```

```python
numero = (df['minutos'] >= 50).sum()

print(f'El Numero de personas que usaron la bicicleta por mas de 50 minutos es: {numero}')
```

```python
#Calcular el tiempo promedio de viaje
vieaje_promedio = df['minutos'].mean()

print(f'El tiempo promedio de viaje es: {vieaje_promedio:.2f} minutos')
```
